### **Simple Graph**

A Simple Graph is a graph that does not contain any loops or parallel edges. In the example below there are 5 vertices and 3 edges. Vertices ${1, 2, 3}$ are connected by edges, while vertex ${4}$ in itself does not contain any edges.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4])

G.add_edges_from([(1, 2), (2, 3)])

nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

plt.title('Simple Graph')
plt.show()
```

### **Directed Graphs (Digraphs)**

Directed Graphs are essentially graphs that contain directed edges which are edges that point to or are directed to a certain vertex. In the graph below contais a directed graph wherein the edges in a graph have directions associated to them. For example, vertex ${1}$ is pointing to vertex ${2}$ and vertex ${2}$ is pointing to vertex ${3}$.

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.DiGraph()

G.add_nodes_from([1,2,3,4,5])
edgelist = [(1,2),(2,3),(3,1),(1,4),(4,5),(5,1)]
G.add_edges_from(edgelist)
nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

plt.title("Directed Graph")
plt.show()
```

### **Complete Graphs**

A Complete Graph is a graph wherein each vertex in a graph is connected to every other vertex inside a graph. For example, vertex ${2}$ here is connected to vertices ${3, 1, 4, 0}$. The same goes for every vertex.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.complete_graph(5)

nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

plt.title('Complete Graph')
plt.show()
```

### **Subgraphs**

A Subgraph is a smaller graph that consists of a subset of the vertices and edges from a larger graph. The graph shown here is a subgraph of the complete graph example shown above.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.complete_graph(5)

edges_to_remove = [(0, 1), (1, 2), (2, 3)]
G.remove_edges_from(edges_to_remove)

nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

plt.title('Subgraph of Complete Graph')
plt.show()
```

### **Bipartite Graphs**

A Bipartite Graph usually contains two sets of vertices. Any vertex within a given set should not be adjacent to any vertex within the same set. In the example below, we have two sets:

Set 1 = {1, 2, 3}

Set 2 = {A, B, C}

No vertex from Set 1 is adjacent to another vertex from Set 1 and no vertex from Set 2 is adjacent to another vertex from Set 2.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from([1, 2, 3], bipartite=0)

G.add_nodes_from(['A', 'B', 'C'], bipartite=1)

G.add_edges_from([(1, 'A'), (1, 'B'), (2, 'B'), (2, 'C'), (3, 'A')])

pos = nx.bipartite_layout(G, [1, 2, 3])
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

plt.title('Bipartite Graph')
plt.show()
```

### **Degree of a Vertex**

The degree of a vertex is defined as the count of edges that connect to that vertex. If a vertex has a loop (an edge connecting it to itself), it contributes 2 to its degree.

deg($v_1$) = 3

deg($v_2$) = 2

deg($v_3$) = 2

deg($v_4$) = 2 (loops are counted as 2)

deg($v_5$) = 0

deg($v_6$) = 1

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5, 6])
G.add_edges_from([(1, 2), (1, 3), (3,2), (1,6), (4,4)])

nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

plt.title('Degrees of a vertex')
plt.show()
```

### **Travelling in a Graph**

To travel within a graph, you move from one vertex to another by following a sequence of adjacent edges.

- A $\textbf{walk from A to D}$ is A  B  C  D
- A $\textbf{trail}$ could be any walk that does not contain a repeated edge.
- A $\textbf{path}$ is a walk that does not contain a repeated vertex.
- A $\textbf{closed walk}$ is a walk  from $v$ to $w$ that starts and ends in the same vertex.
- A $\textbf{circuit}$ is essentially a closed walk that does not contain any repeated edges.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

nodes = [1, 2, 3, 4]
G.add_nodes_from(nodes)

edges = [(1, 2), (2, 3), (3, 4), (4, 1)]
G.add_edges_from(edges)


nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')
plt.show()
```

### **Connectedness as a Graph**


The given graph is connected because there exists a walk from vertex ${1}$ to vertex ${4}$.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4)])

is_connected = nx.is_connected(G)

nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

if is_connected:
    plt.title('Connected Graph')

plt.show()
```

### **Euler's Circuit**

An Eulerian Circuit is a closed walk in a graph that visits every edge exactly once and returns to the starting vertex. Such a circuit exists in a graph if and only if the graph is connected and every vertex has an even degree.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (4, 1), (3, 4)])

is_connected = nx.is_connected(G)

nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

if is_connected:
    plt.title("Euler's Circuit")

plt.show()
```

### **Euler's Trail**

A Eulerian trail, also known as an Eulerian path, is a walk in a graph that traverses every edge exactly once. Unlike an Eulerian circuit, a Eulerian trail may start and end at different vertices and is not required to be a closed loop.

In the graph below, a Euler's Path can be seen that starts with vertex ${3}$ and vertex ${2}$.

The trail starts with 3  1  2  4  3  2.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['1','2','3','4'])
G.add_edges_from([('1', '2'), ('1', '3'), ('2', '3'), ('3', '4'), ('2','4')])


pos = {'1': (1, 2), '2': (0, 0), '3': (2, 0), '4': (1, 1)}
nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

plt.title("Euler's Trail")
plt.show()
```

### **Hamiltonian Circuits**


A Hamiltonian circuit is a closed walk in a graph that visits every vertex exactly once, except for the starting vertex which is visited twice. In other words, it is a circuit that visits every vertex exactly once, while also ending at the same vertex it started in.

Circuit: 2  4  5  1  3  2

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()

G.add_nodes_from([1,2,3,4,5])
edgelist = [(1,2),(2,3),(3,1),(1,4),(4,5),(5,1),(2,4)]
G.add_edges_from(edgelist)
pos=nx.planar_layout(G)
nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12, font_weight='bold')

plt.title("Hamiltonian Circuit")
plt.show()
```

### **Travelling Salesman Problem**


The Traveling Salesman Problem is about finding the shortest possible route that visits each vertex exactly once and returns to the starting point, given a graph with $n$ vertices where each edge is associated with a distance. This problem aims to minimize the total distance traveled.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from([1,2,3,4])
G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(2,4),(3,4)])

pos = nx.spring_layout(G)
nx.draw(G, pos=pos, with_labels=True)

edge_labels = {(1,2): 15,(1,3): 10,(1,4): 5,(2,3): 25,(2,4): 50,(3,4): 20}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, label_pos=0.4)
plt.show()
```
